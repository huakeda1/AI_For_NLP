## 1. 获取歌曲列表
- 这里获取的是首页发现->歌单中的列表；
- 取列表的第一页（后序可扩展到制定页数）；
- 获取歌单中的所有歌曲；

```python
import requests, re
```

```python
def get_response(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_3) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.0.3 Safari/605.1.15'}
    return requests.get(url, headers=headers).content
```

```python
url = 'https://music.163.com/discover/playlist'
get_response(url)
```

```python
def get_playlist():
    list_pattern = re.compile('/playlist\?id=\d+')
    all_playlist_str = ''.join(list_pattern.findall(str(get_response(url))))
    id_pattern = re.compile('\d+')
    all_id_playlist = id_pattern.findall(all_playlist_str)
    return all_id_playlist
```

```python
get_playlist()
```

```python
def get_song_list(playlist_id):
    url = 'https://music.163.com/playlist?id=' + playlist_id
    song_pattern = re.compile('/song\?id=\d+')
    all_song_str = ''.join(song_pattern.findall(str(get_response(url))))
    song_id_pattern = re.compile('\d+')
    return song_id_pattern.findall(all_song_str)    
```

```python
get_song_list('2509086690')
```

```python
def get_all_songs():
    play_list = get_playlist()
    all_songs = []
    for play in play_list:
        all_songs.extend(get_song_list(play))
    return all_songs
```

```python
all_songs = get_all_songs()
```
